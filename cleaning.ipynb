```python
import pandas as pd
import numpy as np
```

Import dataset

```python
original_data=pd.read_csv('mkbhd.csv')
```

```python
df=original_data.copy()
```

```python
df
```

Drop unnecessary columns 

```python
df.columns
```

```python
df=df.drop(['Unnamed: 0','Duration'],axis=1)
```

```python
df
```

```python
df.dtypes
```

Drop na 

```python
df=df.dropna()
```

Fix views column by removing views text and removing columns and make it to integer

```python
def views_helper(n):
    n=n[:-6]
    n=n.replace(',','')
    return int(n)
```

There are some null values

```python
newviews=[]
for i in df['Views']:
    try:
        t=views_helper(i)
    except:
        t=0
    newviews.append(t)
len(newviews)
```

```python
df['Views']=newviews
```

```python
df['Views']
```

```python
df
```

Fix Comments column

```python
#Comments column name has a type, lets fix it
df.rename(columns={'Commments':'Comments'},inplace=True)
df
```

Remove the commas and fix the dtype

```python
df['Comments']=df['Comments'].str.replace(',','')
```

```python
def toint(n):
    if n.isnumeric():
        return int(n)
    return -1
```

```python
df['Comments']=df['Comments'].apply(lambda x:toint(x))
```

```python
df
```

Lets fix Likes column

```python
df['Likes']=df['Likes'].apply(lambda x:x[:-6]).apply(lambda x:x.replace(',','')).apply(lambda x:int(x))
```

```python
df
```

Finally, lets fix data column

```python
def datecolumn(n):
    if 'Pre' in n:
        return n[10:]
    if 'Stre' in n:
        return n[17:]
    else:
        return n
```

```python
df['Date']=pd.to_datetime(df['Date'].apply(lambda x:datecolumn(x)))
```
